Converting string values to list

```python
def convert_to_str(str):
  return eval(str)
```

```python
df.loc[~df['genres'].isna(),'genres']=df.loc[~df['genres'].isna(),'genres'].apply(convert_to_str)
```

```python
df.loc[~df['cast'].isna(),'cast']=df.loc[~df['cast'].isna(),'cast'].apply(convert_to_str)
```

```python
df.loc[~df['crew'].isna(),'crew']=df.loc[~df['crew'].isna(),'crew'].apply(convert_to_str)
```

```python
#creating a list of genres
genres_list=[]

#iterating over each observation in the data frame
for index, row in genres_no_na.iterrows():
#identify the list of genres
  genres_list=row['genres']
#iterate over list of dic
  for elem in genres_list:
    genres_list.append(elem['name'])
```
